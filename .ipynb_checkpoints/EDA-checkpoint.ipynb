## Histograms

```python
import math

def plot_hist(df, variable, n_bins, round_edges=True, graph_size=None): # default size
    fig, ax = plt.subplots(figsize=graph_size)
    plt.title("Histogram of {}".format(variable))
        
    if round_edges:
        val_range = df[variable].max() - df[variable].min()
        step = round(val_range/n_bins)
            
        round_min = math.floor(df[variable].min())
        round_max = math.ceil(df[variable].max())
        bins = [edge for edge in range(round_min, round_max+step, step)]
    
        ax.set_xticks(bins)
        ax.hist(df[variable].dropna(), bins=bins)
        
    else:
        ax.hist(df[variable].dropna(), bins=n_bins)

    
plot_hist(foodie_df, 'Average_Cost', 15, graph_size=(7, 5))
```

**Comment:** Most restaurants cost less than 50 dollars on average with over 1000 restaurants (about half of the total) having an average cost between 15 and 25 (this makes sense since the mean is about 20). In other words, a majority of the restaurants are affordable while only a few are luxurious (the number of restaurants decreases exponentially as price goes up).

```python
plot_hist(foodie_df, 'Minimum_Order', 20, graph_size=(11, 5))
```

**Comment:** The vast majority of restaurants have there minimum orders between 44 and 66 dollars (makes sense since the mean is about 53).

```python
plot_hist(foodie_df, 'Numeric_Rating', 8, round_edges=False, graph_size=(7, 5))
```

**Comment:** This looks like a normal distribution centered around 3.6 (the mean). However, there are more higher rated restaurants than lower rated ones (the right tail is a bit higher).

```python
plot_hist(foodie_df, 'Votes', 100, round_edges=False, graph_size=(12, 5))
```

**Comment:** Most restaurants have only a few votes but there're some of them which receive a huge number of votes (outliers). This will be clearer if we look at the Box plot.

This indicates as we climb up the "ladder", the gap becomes wider. There are also many outliers detected.

```python
plot_hist(foodie_df, 'Reviews', 100, round_edges=False, graph_size=(12, 5))
```

```python
plt.figure(figsize=(15, 5))
sns.boxplot(foodie_df.Reviews)
plt.title('Boxplot of Reviews')
```

# Pairwise exploration

```python
# Next, we look at the correlation matrix and the relationships between the variables:
corr_df = foodie_df.loc[:, 'Average_Cost':].corr()
corr_df
```

```python
def plot_corr_mat(corr_df, title=None):
    fig, ax = plt.subplots(figsize=(11, 11))
  
    mat_img = ax.matshow(corr_df, cmap=plt.cm.binary)
    fig.colorbar(mat_img) 

    ax.set_xticks(range(len(corr_df.columns)))
    ax.set_xticklabels(corr_df.columns)
    ax.set_yticks(range(len(corr_df.columns)))
    ax.set_yticklabels(corr_df.columns)

    plt.title(title, size=20)
    plt.show()
    
abs_corr_df = corr_df.applymap(lambda x: abs(x))
plot_corr_mat(abs_corr_df, 'Pearson Correlation')
```

**Comment:** 
* ``Votes`` and ``Reviews``, as expected, are strongly correlated. They are also correlated to ``Average_Cost`` though the correlation is not as strong. 
* ``Rating`` is correlated to ``Average_Cost``, ``Votes``, and ``Reviews``. That the correlation with ``Average_Cost`` is positive suggests that higher cost on average actually indicates a better rating.
* ``Cook_Time`` is most correlated to ``Average_Cost`` and ``Minimum_Order`` but this correlation is also very mild.

```python
spm_corr_df = foodie_df.loc[:, 'Average_Cost':].corr('spearman')
abs_spm_corr_df = spm_corr_df.applymap(lambda x: abs(x))
plot_corr_mat(abs_spm_corr_df, 'Spearman Correlation')
spm_corr_df
```

**Comment:** 
* ``Rating``'s correlations with ``Votes`` and ``Reviews`` are significantly stronger when we use the Spearman method. This suggests that the relationships are not linear.
* More importantly, ``Cook_Time_Mins`` is also more correlated to Votes and Reviews if the Spearman method is used. This implies that the relationships are not linear.

## A new feature: number of cuisines?

## What do people use coffee with?

```python
from collections import defaultdict 

def what_goes_with(cuisine_exp):
    cuisines_records = foodie_df.loc[foodie_df.Cuisines.str.contains(cuisine_exp)].Cuisines
    go_with = defaultdict(int)
    
    for cuisines_info in cuisines_records.values:
        cuisines_list = cuisines_info.split(',')
        for cuisine in cuisines_list:
            cuisine = cuisine.strip()
            if cuisine not in cuisine_exp:
                go_with[cuisine] += 1
                
    return go_with
                
go_with_coffee = what_goes_with('Coffee|Cafe')

# because I make no change to the dictionary go_with_coffee: their keys() and values() will match:
go_with_coffee_df = pd.DataFrame({'Cuisine': list(go_with_coffee.keys()), 'Count': list(go_with_coffee.values())})
go_with_coffee_df.nlargest(10, columns=['Count'])
```

**Comment:** We see that coffee is mostly served with Fast Food, other beverages, and Italian as shown above. You can try this with other cuisines, too.
